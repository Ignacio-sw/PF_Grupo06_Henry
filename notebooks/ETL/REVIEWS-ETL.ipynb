Hay un total de 611 archivos para leer y procesar, eliminando los registros que contengan un gmap_id no coincidente con la categoría elegida

```python
lenths = []
filas_encontradas=[]
for path in paths:
    df = pd.read_json(path, lines=True, encoding='utf-8')
    df = df.merge(sitios, how='inner', on='gmap_id')
    
    lenght = df.shape[0]
    print(f'Archivo: {path}\n lenght: {lenght}')
    if lenght > 0:
        filas_encontradas.append(df)
    #lenths.append(lenght)
df_consumo_final=pd.concat(filas_encontradas, ignore_index=True)


# Eliminar columnas innecesarias
columnas_a_eliminar = ["time", "pics", "resp"]
df_consumo_final.drop(columns=columnas_a_eliminar, inplace=True, errors='ignore')
df_consumo_final["text"]=df_consumo_final["text"].map(lambda txt: txt.strip(), na_action="ignore").fillna("No comment")
df_consumo_final["text"]=df_consumo_final["text"].map(lambda txt: txt.replace("\n", " "), na_action="ignore")


df_consumo_final.to_csv(r"C:\Users\Usuario\Desktop\GitHub\DataSets\DataWarehouse\Google-CF.csv", index=False)
```
