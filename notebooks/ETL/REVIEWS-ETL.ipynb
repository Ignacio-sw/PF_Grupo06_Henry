```python


# Directorio principal que quieres recorrer
directorio_principal = r".\DataSets\REVIEW-ESTADOS\reviews-estados"

# Recorrer cada carpeta y archivo dentro del directorio principal
for carpeta_actual, carpetas, archivos in os.walk(directorio_principal):
    print(f"Carpeta actual: {carpeta_actual}")

    # Lista para almacenar los datos de los archivos JSON en esta carpeta
    datos_json = []

    # Recorrer los archivos dentro de la carpeta actual
    for archivo in archivos:
        ruta_completa_archivo = os.path.join(carpeta_actual, archivo)
        print(f"Archivo encontrado: {ruta_completa_archivo}")

        # Verificar si el archivo es un archivo JSON
        if archivo.endswith(".json"):
            # Abrir el archivo JSON y leer cada línea
            with open(ruta_completa_archivo, 'r') as archivo_json:
                for linea in archivo_json:
                    try:
                        # Cargar cada línea como un objeto JSON y agregarlo a la lista
                        datos_json.append(json.loads(linea))
                    except json.JSONDecodeError as e:
                        print(f"Error al decodificar JSON en {ruta_completa_archivo}: {e}")

    # Guardar todos los datos en un archivo JSON separado por cada carpeta
    ruta_archivo_salida = os.path.join(carpeta_actual, "datos_concatenados.json")
    with open(ruta_archivo_salida, 'w') as archivo_salida:
        json.dump(datos_json, archivo_salida)

    print(f"Los datos se han guardado en el archivo: {ruta_archivo_salida}")

```

```python


# Directorio principal que quieres recorrer
directorio_principal = r".\DataSets\REVIEW-ESTADOS\reviews-estados"
# DataFrame para almacenar los datos concatenados
df_data_sitios = pd.read_csv(r".\DataSets\DataWarehouse\data_sitios.csv")
# Recorrer cada carpeta y archivo dentro del directorio principal
for carpeta_actual, carpetas, archivos in os.walk(directorio_principal):
    print(f"Carpeta actual: {carpeta_actual}")

    # Verificar si hay un archivo datos_concatenados.json en la carpeta actual
    ruta_datos_concatenados = os.path.join(carpeta_actual, "datos_concatenados.json")
    if os.path.isfile(ruta_datos_concatenados):
        print(f"Archivo encontrado: {ruta_datos_concatenados}")

        # Cargar el archivo JSON en un DataFrame
        try:
            df_concatenados = pd.read_json(ruta_datos_concatenados)
        except pd.errors.EmptyDataError:
            print(f"El archivo {ruta_datos_concatenados} está vacío.")
            continue

        # Eliminar columnas innecesarias
        columnas_a_eliminar = ["time", "pics", "resp"]
        df_concatenados.drop(columns=columnas_a_eliminar, inplace=True, errors='ignore')
        # Unir el DataFrame actual con el DataFrame concatenado
        df_combinado = pd.merge(df_concatenados, df_data_sitios, on='gmap_id')

        # Guardar los datos en un archivo CSV
        ruta_archivo_salida = os.path.join(carpeta_actual, "datos_concatenados.csv")
        df_concatenados.to_csv(ruta_archivo_salida, index=False)
        
        print(f"Los datos se han guardado en el archivo: {ruta_archivo_salida}")
    else:
        print("No se encontró el archivo datos_concatenados.json en esta carpeta.")


```

```python
# Directorio principal que quieres recorrer
directorio_principal = r".\DataSets\REVIEW-ESTADOS\reviews-estados"

# DataFrame para almacenar los datos concatenados
df_data_sitios = pd.read_csv(r".\DataSets\DataWarehouse\data_sitios.csv")

# Recorrer cada carpeta y archivo dentro del directorio principal
for carpeta_actual, carpetas, archivos in os.walk(directorio_principal):
    print(f"Carpeta actual: {carpeta_actual}")

    # Recorrer los archivos dentro de la carpeta actual
    for archivo in archivos:
        # Verificar si el archivo es un archivo CSV
        if archivo.endswith(".csv"):
            ruta_completa_archivo = os.path.join(carpeta_actual, archivo)
            print(f"Archivo encontrado: {ruta_completa_archivo}")

            # Cargar el archivo CSV en un DataFrame
            try:
                df_archivo = pd.read_csv(ruta_completa_archivo)
            except pd.errors.EmptyDataError:
                print(f"El archivo {ruta_completa_archivo} está vacío.")
                continue

            # Unir el DataFrame actual con el DataFrame concatenado
            df_combinado = pd.merge(df_archivo, df_data_sitios, on='gmap_id')
            
            # Guardar el DataFrame combinado en un archivo CSV en la misma carpeta
            nombre_archivo_salida = f"df_combinado_{archivo}"
            ruta_archivo_salida = os.path.join(carpeta_actual, nombre_archivo_salida)
            df_combinado.to_csv(ruta_archivo_salida, index=False)
            
            print(f"Los datos combinados se han guardado en el archivo: {ruta_archivo_salida}")


```
